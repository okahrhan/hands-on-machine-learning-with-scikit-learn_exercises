```python
import tensorflow as tf
from tensorflow import keras
```

```python
import pandas as pd
import numpy as np
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train_full.shape
```

```python
X_train_df = pd.DataFrame(X_train_full.reshape(X_train_full.shape[0], -1))
y_train_df = pd.DataFrame(y_train_full, columns=['label'])
X_test_df = pd.DataFrame(X_test.reshape(X_test.shape[0], -1))
y_test_df = pd.DataFrame(y_test, columns=['label'])

display(X_train_df.head())
display(y_train_df.head())
display(X_test_df.head())
display(y_test_df.head())
```

```python
X_valid , X_train = X_train_full[:5000]/ 255.0, X_train_full[5000:]/255.0
y_valid , y_train = y_train_full[:5000], y_train_full[5000:]
X_test = X_test/255.0
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
"Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(300, activation='relu'))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10 , activation='softmax'))
```

```python
model = keras.models.Sequential([
 keras.layers.Flatten(input_shape=[28, 28]),
 keras.layers.Dense(300, activation="relu"),
 keras.layers.Dense(100, activation="relu"),
 keras.layers.Dense(10, activation="softmax")
])
```

```python
model.summary()
```

```python
model.layers
```

```python
hidden1 = model.layers[1]
hidden1.name
```

```python
weights, biases = hidden1.get_weights()
```

```python
weights
```

```python
biases
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=30, validation_data=(X_valid, y_valid))
```

```python
X_new = X_test[:3]
y_proba = model.predict(X_new)
y_proba.round(2)
```

```python
y_proba = model.predict(X_new)
y_pred = np.argmax(y_proba, axis=1)
y_pred
```
